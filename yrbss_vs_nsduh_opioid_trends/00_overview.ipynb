# YRBSS vs NSDUH Opioid Misuse Trends

This project compares adolescent opioid misuse trends between:

- **NSDUH** (national household survey; ages 12–17 subset)
- **YRBSS** (school-based survey; high school students)

The goal is to assess whether both surveillance systems show similar direction and timing of changes in youth opioid misuse over time.

## Notebook map

1. **`01_download_yrbss.ipynb`**
   - Downloads and harmonizes YRBSS opioid misuse indicators from CDC data APIs.
   - Saves cleaned annual prevalence estimates to `data/derived/yrbss/`.

2. **`02_compare_with_nsduh.ipynb`**
   - Downloads NSDUH detailed table estimates for adolescent prescription pain reliever misuse.
   - Joins NSDUH and YRBSS trends and generates comparison figures/tables.

## Outputs

- Harmonized comparison dataset: `data/derived/yrbss_nsduh/yrbss_nsduh_opioid_trends.csv`
- Summary statistics table: `data/derived/yrbss_nsduh/summary_metrics.csv`
- Figures are displayed inline in notebook 02.

## Notes on interpretation

- YRBSS is biennial and school-based; NSDUH is annual and household-based.
- Definitions differ slightly (e.g., YRBSS asks about lifetime non-medical prescription pain medicine use in many years, while NSDUH provides past-year misuse estimates).
- We therefore focus on **trend direction** and **relative change** rather than exact prevalence equivalence.
