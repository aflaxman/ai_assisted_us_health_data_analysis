# YRBSS vs NSDUH Opioid Misuse Trends

This project compares adolescent opioid misuse trends between:

- **NSDUH** (national household survey; ages 12–17 subset)
- **YRBSS** (school-based survey; high school students)

The goal is to assess whether both surveillance systems show similar direction and timing of changes in youth opioid outcomes over time.

## Notebook map

1. **`01_download_yrbss.ipynb`**
   - Downloads/harmonizes YRBSS opioid indicators from CDC API (with local fallback).
   - Writes cleaned trends to `data/derived/yrbss/`.

2. **`02_compare_with_nsduh.ipynb`**
   - Builds cross-survey trend panel.
   - Compares YRBSS lifetime prescription misuse to NSDUH past-year misuse and (if available) NSDUH lifetime prescription misuse from PUF-based extracts.

3. **`03_individual_level_data_check_and_puf_bridge.ipynb`**
   - Documents what YRBSS and NSDUH individual-level public data are available.
   - Attempts PUF-based lifetime Rx misuse extraction for NSDUH ages 12–17 when raw files are present locally.

## Outputs

- `data/derived/yrbss_nsduh/yrbss_nsduh_opioid_trends.csv`
- `data/derived/yrbss_nsduh/summary_metrics.csv`
- `data/derived/yrbss_nsduh/nsduh_lifetime_rx_misuse_12_17_from_puf.csv` (only when local NSDUH PUF data are available)

## Measure scope clarifications

- **YRBSS** commonly provides **lifetime** indicators for heroin use and prescription pain medicine misuse.
- YRBSS public indicators typically do **not** provide DSM-style opioid use disorder/dependence diagnosis variables.
- **NSDUH** PUF microdata can support richer definitions when raw files are available locally.
- Because YRBSS is school-based and NSDUH is household-based, absolute prevalence values are not directly interchangeable.
