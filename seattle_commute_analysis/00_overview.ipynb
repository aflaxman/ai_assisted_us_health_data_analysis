# Seattle Commute Mode & Travel Time Trends

This project uses **American Community Survey (ACS) Public Use Microdata Sample (PUMS)**
data to analyze how Seattle-area workers commute to work and how those patterns have
changed over time.

## Research questions

1. How many Seattle workers use each commute mode (drive alone, bus, bike, walk, etc.)
   and how have these numbers changed from 2013 to 2023?
2. What fraction of Seattle workers use each mode, and how have shares shifted?
3. How long does it take to get to work by car, bus, bike, etc., and have travel times
   changed over time?

## Data

- **ACS 1-Year PUMS** person-level microdata for Washington state (2013–2019, 2021–2023)
- **PUMA reference files** for mapping geographic areas to Seattle
- 2020 ACS data excluded (experimental estimates due to COVID-19 low response rates)

## Key caveats

- **PUMA geography changed** between 2021 and 2022 (2010-based → 2020-based PUMAs).
  The set of "Seattle" PUMAs differs across these vintages, which may cause a
  discontinuity at the 2021–2022 boundary.
- **Work-from-home workers** are classified by POWPUMA at their residence, not their
  employer's location. The WFH fraction for downtown employers may be undercounted
  if the worker lives outside the downtown PUMA.
- **2020 is missing** due to the Census Bureau's experimental data quality for that year.
- Standard errors use the ACS successive difference replication method (80 replicate weights).

## Key results

*(Results from notebook 02, using ACS PUMS 2013–2023.)*

**2023 mode shares (Seattle-area workplace POWPUMA):**
- Drove alone: 57.1%
- Worked from home: 21.4%
- Carpooled: 8.2%
- Bus: 6.0%
- Walked: 4.0%

**Notable trends:**
- Work-from-home surged from ~6% (2019) to ~34% (2021) during COVID, settling at ~21% by 2023
- Drove-alone share declined from ~60–65% pre-pandemic to ~57% in 2023
- Bus ridership dropped during COVID and has not fully recovered
- Bus commuters have the longest average commute (~41 min); walkers the shortest (~17 min)
