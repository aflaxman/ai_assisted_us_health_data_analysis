# NSDUH Opioid Prevalence Estimates

Prevalence of opioid use disorder (OUD) and any opioid use in the US,
stratified by age, sex, race/ethnicity, and year.

**Data source:** National Survey on Drug Use and Health (NSDUH), SAMHSA

## Notebooks

| # | Notebook | Description |
|---|----------|-------------|
| 01 | [01_samhsa_tables.ipynb](01_samhsa_tables.ipynb) | **Path A**: Parse pre-computed SAMHSA detailed tables (2021–2024) with demographic OUD/opioid use estimates |
| 02 | [02_download_pufs.ipynb](02_download_pufs.ipynb) | **Path B step 1**: Download public use files (2002–2023, 22 years) |
| 03 | [03_puf_analysis.ipynb](03_puf_analysis.ipynb) | **Path B step 2**: Compute survey-weighted prevalence from PUF microdata with paired jackknife SEs |

## Outcomes

1. **Opioid Use Disorder (OUD)** — past-year
   - 2002–2014: DSM-IV opioid abuse or dependence (ABODHER, ABODANL)
   - 2015+: DSM-5 OUD (UD5OPIANY / UDPYOPI)
2. **Any Opioid Use** — past-year heroin use OR prescription opioid misuse/use
   - 2002–2014: heroin use OR any pain reliever use (HERYR | ANLYR) — broader definition
   - 2015–2020: opioid misuse (OPINMYR)
   - 2021+: any opioid use including as prescribed (OPIANYYR) — broadest definition

## Stratifications

- **Age**: 12–17, 18–25, 26–34, 35–49, 50–64, 65+ (2005+); 12–17, 18–25, 26–34, 35+ (2002–2004)
- **Sex**: Male, Female
- **Race/Ethnicity**: NH White, NH Black, NH Native Am/AK, NH Asian, NH Multiracial, Hispanic
- **Year**: 2002–2023 (Path B) or 2021–2024 (Path A)

## Comparability Notes

| Period | Methodology | Comparability |
|--------|-------------|---------------|
| 2002–2014 | In-person only, DSM-IV | Comparable within period |
| 2015–2019 | In-person, partial redesign, DSM-5 from 2015 | Break with pre-2015 for some measures |
| 2020 | Disrupted by COVID (Q1+Q4 only) | **Not recommended for trend analysis** |
| 2021–present | In-person + web, DSM-5 | Comparable within period; **NOT comparable to pre-2021** |
