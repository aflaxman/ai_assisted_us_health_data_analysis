# Fibrosis Progression: Data Strategy Investigation

This project investigates publicly available data sources for estimating
**age-, sex-, and race-specific fibrosis progression rates**, suitable for
use in a microsimulation model.

## Notebooks

| # | Notebook | Description |
|---|---------|-------------|
| 01 | `01_fibrosis_progression_data_strategy.ipynb` | Literature review, data source comparison, and NHANES 1999–2018 FIB-4 pseudo-cohort analysis (10 cycles) |
| 02 | `02_lsm_analysis.ipynb` | LSM (liver stiffness, kPa) analysis using NHANES FibroScan data (2017–18 + 2021–22); FIB-4 vs LSM concordance; EPIC Cosmos data strategy |

## Key Findings

### FIB-4 pseudo-cohort (notebook 01)
- Annual FIB-4 increase: ~0.044 (male) / 0.047 (female) per year of aging (NHANES pseudo-cohort)
- Non-Hispanic White males show the steepest age-related FIB-4 increase (~0.048/yr)
- FIB-4 > 2.67 prevalence rises from ~5% at 35–39 to ~25–35% at 75+

### LSM analysis (notebook 02)
- NHANES 2021–22 (cycle L) confirmed to include FibroScan data (LUX_L.xpt; n=7,199)
- LSM ≥ 9.5 kPa (F3–F4) prevalence: 3–13% depending on age/sex/race
- Two-cycle pseudo-cohort (4-year gap): annual LSM change ~0.0–0.07 kPa/year
- FIB-4 and LSM show moderate correlation (Spearman r ≈ 0.3–0.5); correlation weakens in older age groups where FIB-4 is confounded by age-related platelet decline
- Steatosis (CAP ≥ 280 dB/m, S3) strongly associated with elevated LSM at all ages

### EPIC Cosmos potential (notebook 02, section 8)
- True individual-level FIB-4 trajectories from repeated EHR labs
- Enables Markov transition model: P(Low→Interm), P(Interm→High), P(regression) by age/sex/race/DM/BMI
- Key limitation: healthcare-seeking population (not nationally representative)
