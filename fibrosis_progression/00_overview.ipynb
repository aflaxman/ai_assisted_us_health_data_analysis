# Fibrosis Progression: Data Strategy Investigation

This project investigates publicly available data sources for estimating
**age-, sex-, and race-specific fibrosis progression rates**, suitable for
use in a microsimulation model.

## Notebooks

| # | Notebook | Description |
|---|---------|-------------|
| 01 | `01_download_nhanes_lsm.ipynb` | Download NHANES FibroScan cycles (2017–2020 pre-pandemic, 2021–22) → parquets in `../data/derived/`. Run once. |
| 02 | `02_fib4_pseudo_cohort.ipynb` | Literature review, data source comparison, NHANES 1999–2018 FIB-4 pseudo-cohort (10 cycles, age/sex/race progression rates). Depends on `nhanes_mortality_fibrosis/01_data_download.ipynb`. |
| 03 | `03_lsm_analysis.ipynb` | LSM (liver stiffness) analysis using FibroScan data (2017–2020 + 2021–22); FIB-4 vs LSM concordance; EPIC Cosmos data strategy. Depends on notebook 01. |

## Key Findings

### FIB-4 pseudo-cohort (notebook 02)
- Annual FIB-4 increase: ~0.044 (male) / 0.047 (female) per year of aging (NHANES pseudo-cohort)
- Non-Hispanic White males show the steepest age-related FIB-4 increase (~0.048/yr)
- FIB-4 > 2.67 prevalence rises from ~5% at 35–39 to ~25–35% at 75+

### LSM analysis (notebook 03)
- Three NHANES LSM cycles available: 2017-18 (J), 2017-2020 pre-pandemic (P_), 2021-22 (L)
- The 2017-2020 pre-pandemic cycle supersedes 2017-18 and includes Jan 2019–Mar 2020 data
- LSM ≥ 9.5 kPa (F3–F4) prevalence: 3–13% depending on age/sex/race
- Two-window pseudo-cohort (2017-2020 → 2021-22, ~3.5-year gap): annual LSM change ~0–0.07 kPa/year
- FIB-4 and LSM show weak overall correlation (Spearman r ≈ 0.05) in general population; FIB-4 is heavily confounded by age-related platelet decline independent of fibrosis
- Severe steatosis (CAP S3, ≥280 dB/m) strongly associated with elevated LSM at all ages

### EPIC Cosmos potential (notebook 03, section 8)
- True individual-level FIB-4 trajectories from repeated EHR labs
- Enables Markov transition model: P(Low→Interm), P(Interm→High), P(regression) by age/sex/race/DM/BMI
- Key limitation: healthcare-seeking population (not nationally representative)
